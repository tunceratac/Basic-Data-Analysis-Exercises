```python
# Define the Titanic dataset from the Seaborn library
import seaborn as sns
df = sns.load_dataset("titanic")
df.head()
```

```python
# Find the number of male and female passengers in the Titanic dataset.
df["sex"].value_counts()
```

```python
# Find the number of unique values for each column
df.nunique()
```

```python
# Find the number of unique values of the variable pclass.
df["pclass"].nunique()
```

```python
# Find the unique values of the variable pclass
df["pclass"].unique()
```

```python
# Check the type of the embarked variable. 
df["embarked"].dtype
```

```python
# Change its type to category and check again.
df["embarked"] = df["embarked"].astype("category")
df["embarked"].dtype
```

```python
# Show all information for those with an embarked value of C.
df[df["embarked"] == "C"].head()
```

```python
# Show all information for those whose embarked value is not S.
df[df["embarked"] != "S"].head()
```

```python
# Show all information for female passengers younger than 30 years old
df[(df["age"] < 30) & (df["sex"] == "female")].head()
```

```python
# Show information for passengers with a Fare over 500 or Age over 70 years of age.
df[(df["fare"] > 500) |  (df["age"] > 70)].head()
```

```python
# Find the sum of the null values in each variable.
df.isnull().sum()
```

```python
# Remove who variable from dataframe
df.drop("who", axis=1, inplace=True)
df.head()
```

```python
# Fill in the empty values in the deck variable with the most repeated value (mode) of the deck variable
df["deck"].fillna(df["deck"].mode()[0], inplace=True)
df.isnull().sum()
```

```python
df.head()
```

```python
# Fill in the blank values in the age variable with the median of the age variable
df["age"].fillna(df["age"].median(), inplace=True)
df.isnull().sum()
```

```python
df.head()
```

```python
# Find the sum, count, mean values of the pclass and sex variables of the survived variable.
df.groupby(["sex", "pclass"]).agg({"survived": ["sum", "count", "mean"]})
```

```python
# Write a function that will return 1 for those under 30, 0 for those equal to or above 30. 
# Using the function you wrote, create a variable named age_flag in the titanic data set. (use apply and lambda constructs)
def func(age):
    if age < 30:
        return 1
    else:
        return 0
df["age_flag"] = df["age"].apply(lambda x: func(x))
df.head()
```

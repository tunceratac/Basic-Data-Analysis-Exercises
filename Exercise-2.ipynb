```python
# Define the Tips dataset from the Seaborn library
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

```python
# Find the sum, min, max and mean values of the total_bill value 
# according to the categories (Dinner, Lunch) of the time variable.
df.groupby("time").agg({"total_bill": ["sum", "min", "max", "mean"]})
```

```python
# Find the sum, min, max and mean values of the total_bill values according to day and time.
df.groupby(["day", "time"]).agg({"total_bill": ["sum", "min", "max", "mean"]})
```

```python
# Find the sum, min, max and mean values of the total_bill and 
# tip values of the lunch time and female customers according to the day.
df[(df["time"] == "Lunch") & (df["sex"] == "Female")].groupby("day").agg({"total_bill": ["sum","min","max","mean"],
                                                                           "tip":  ["sum","min","max","mean"]})
```

```python
# What is the average of orders with size less than 3 and total_bill greater than 10? (use loc)
df.loc[(df["size"] < 3) & (df["total_bill"] > 10 ) , "total_bill"].mean()
```

```python
# Create a new variable called total_bill_tip_sum. Let him give the sum of the total bill and tip paid by each customer.
df["total_bill_tip_sum"] = df["total_bill"] + df["tip"]
df.head()
```

```python
#Find the mean of the total_bill variable separately for men and women. 0, above and equal to those below the 
# averages you found create a new total_bill_flag variable where 1 is given to those.

# The averages of Female for women and Male for men will be considered. 
# Start by writing a function that takes sex and total_bill as parameters. (will include if-else conditions)

def func(sex, bill):
    if sex == "Male":
        if bill < df[(df["sex"] == "Male")]["total_bill"].mean():
            return 0
        else:
            return 1
    else:
        if bill < df[(df["sex"] == "Female")]["total_bill"].mean():
            return 0
        else:
            return 1

df["total_bill_flag"] = df[["sex", "total_bill"]].apply(lambda x: func(x["sex"], x["total_bill"]), axis=1)
df.head()
```

```python
# Using the total_bill_flag variable, observe the number of below and above average by sex
df.groupby(["sex","total_bill_flag"]).agg({"total_bill_flag":"count"})
```

```python
# Sort the data from largest to smallest according to the total_bill_tip_sum variable 
# and assign the first 30 people to a new dataframe.
dff = df.sort_values("total_bill_tip_sum", ascending=False)[:30]
dff.head()
```
